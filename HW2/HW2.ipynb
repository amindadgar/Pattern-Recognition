```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1
Plot the probability of x given θ using the formula below. <br>

$p(x|θ) = 
\begin{cases}
    \theta e^{{-\theta  x}},& \text{if } x\geq 0\\
    0,              & \text{otherwise}
\end{cases}
$


### plot the formula for given θ = 1  

```python
SAMPLE_SIZE = 100

## x_1 is the negative side of the plot that y = 0
x_1 = np.linspace(-10, 0, int(SAMPLE_SIZE/2))
y_1 = np.zeros(50)
## x_2 is the positive side of the plot
x_2 = np.linspace(0.01, 10, int(SAMPLE_SIZE/2))
y_2 = np.exp(-x_2)

X = np.concatenate((x_1, x_2), axis = 0)
Y = np.concatenate((y_1, y_2), axis = 0)

## To make the function discontinues
pos = np.where(X==0)[0]
X[pos] = np.nan
Y[pos] = np.nan

plt.plot(X , Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### plot $ p(x|\theta) $ versus $\theta (0 \leq \theta \leq 5)$ for $x = 2$

```python
fig ,axes = plt.subplots(1, 2,figsize=(14, 5))

## plot the last plot
axes[0].plot(X,Y)
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("p(x|theta)")


## plot the second plot
theta = np.linspace(0,5,SAMPLE_SIZE)
y_theta = theta * np.exp(-2 * theta)

axes[1].plot(theta, y_theta)
axes[1].set_xlabel("theta")
axes[1].set_ylabel("Y")
axes[1].set_title("p(x=2| 0 <= theta <= 5) ")

plt.show()
```

### On the upeer left plot mark the Maximum-likelihood estimation for the graph for 𝜃 = 1
By creating the log-likelohood function and making its derivation equal to zero we can find the ML estimation value. <br>
Maximum likelihood value is found as below
$$ \hat \theta = \frac{1}{1/n \sum_{k=1}{n} } $$

```python
def sigma(lower, upper, values):
    """
    The mathematical sigma function, To iterate over a value and return the sum of it
    """
    total = 0
    for i in range(lower, upper):
        total += values
    
    return total
```

```python
plt.plot(theta, y_theta)
theta_hat = 1 / ((1/ SAMPLE_SIZE) * sigma(0, SAMPLE_SIZE, SAMPLE_SIZE))
plt.scatter(1, theta_hat, color='r')
plt.title("p(x=2| 0 <= theta <= 5)")
plt.show()
```
