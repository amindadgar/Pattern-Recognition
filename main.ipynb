```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_height(x,y):
    """
    get the height of x and y
    this function is can be change due to your preference
    """
    return x**2 + 3* (y)


## if you want more create more sets and add it in the dataset array
def plot_contours(dataset):
    """
    plot contours and scatter for many datasets as possible ( no limits! )
    """
    row_count = len(dataset)
    column_count = 2


    ## plot each the data within an ax
    fig, axes = plt.subplots(row_count, column_count, figsize= (8*column_count, 5*row_count))
    for idx, data in enumerate(dataset):
        x = data[:,0]
        y = data[:,1]
        z = get_height(x,y)

        axes[idx][0].tricontour(x,y,z)
        axes[idx][0].set_title("Data Contour")
        axes[idx][1].scatter(x,y,z)
        axes[idx][1].set_title("Data positions")
```

```python
############################ plot Gaussian normal distribution with its contour with numpy ############################
data_size = 500

cov_matrix1 = np.matrix("[4 0; 0 4]")
mean1 = [0,0]
set1 = np.random.multivariate_normal(mean1, cov_matrix1, size=data_size)

cov_matrix2 = np.matrix("[6 0; 0 12]")
mean2 = [5,5]
set2 = np.random.multivariate_normal(mean2, cov_matrix2, size=data_size)
```

```python
dataset = [set1, set2]

plot_contours(dataset)
```
