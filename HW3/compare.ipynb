```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Question 2 Part c
Compare the results from Question 1 and Question 2 <br>
**Note:** that we're comparing A to F dataset.

```python

## read the results from question 1 and questions 2 outputs
q1_results = pd.read_csv('Q1/confusion_matrix.csv')
q2_results = pd.read_csv('Q2/Q2-results.csv')
```

```python

plt.figure(figsize=(12, 7))

plt.title("Comparison between results from KNN algorithm and Bayes classifier")
x = ['Atrain', 'Btrain', 'Ctrain','Dtrain', 'Etrain', 'Ftrain']

plt.plot(x, q1_results[q1_results.K == 1][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 2][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 5][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 10][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 15][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 20][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 50][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 100][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 500][:6].accuracy.values)
plt.plot(q1_results[q1_results.K == 1000][:6].accuracy.values)



plt.plot(q2_results[q2_results.isotropic == False][:6].accuracy.values)
plt.plot(q2_results[q2_results.isotropic == True][:6].accuracy.values)

plt.legend(['KNN with K = 1','KNN with K = 2','KNN with K = 5', 'KNN with K = 10'
                    ,'KNN with K = 15', 'KNN with K = 20', 'KNN with K = 50', 'KNN with K = 100',
                    'KNN with K = 500', 'KNN with K = 1000'
                     ,'Bayes non-isotropic', 'Bayes isotropic'])

plt.ylabel("accuracy")

plt.savefig('comparison.png', )
plt.show()

```
